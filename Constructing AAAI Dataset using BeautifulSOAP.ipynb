```python
#import the necessary modules

import requests
from bs4 import BeautifulSoup
import re
```

```python
page = requests.get('https://aaai.org/Library/AAAI/aaai18contents.php')
soup = BeautifulSoup(page.content, 'html.parser')
papersLinksTags = soup.find_all('p', class_='left')
#papersLinksTags = soup.select("p a")
#print(papersLinks)

papersLinks = []
#for each paper link
for paperLinkTag in papersLinksTags:
    print(paperLinkTag)
    
    paperLink = re.findall(r'<a href="(.*?)">', str(paperLinkTag))
    paperLink = paperLink[0]
    print(paperLink)
    
    papersLinks.append(paperLink)
    print('-------------------------------------------')

print('Done')
```

```python
#open each link in the paperLinks list
#get the actual paper link

actualPapersLinks = []
#for each paper link
for paperLink in papersLinks:
    page = requests.get(paperLink)
    soup = BeautifulSoup(page.content, 'html.parser')
    
    actualPaperLink = soup.select("a")
    print(actualPaperLink)
    actualPaperLink = re.findall(r'<a href="(.*?)">', str(actualPaperLink))
    actualPaperLink = actualPaperLink[0]
    print(actualPaperLink)
    
    actualPapersLinks.append(actualPaperLink)
    print('-------------------------------------------')

print('Done')
```

```python
#Please, choose the directory where the dataset files will be saved.

from  tkinter import *
from tkinter import filedialog
import os

root = Tk()
root.directory = filedialog.askdirectory()
dataPath = root.directory
print (dataPath)
root.destroy()

titleKeywordsWithNoKeyWords       = 'titleKeywordsWithNoKeyWords.txt' 
titleKeywordsWithoutNoKeyWords    = 'titleKeywordsWithoutNoKeyWords.txt' 
abstractTitle                     = 'abstractTitle.txt'
abstractKeywordsWithNoKeyWords    = 'abstractKeywordsWithNoKeyWords.txt'
abstractKeywordsWithoutNoKeyWords = 'abstractKeywordsWithoutNoKeyWords.txt'

titleKeywordsWithNoKeyWords       = os.path.join(dataPath, titleKeywordsWithNoKeyWords)
titleKeywordsWithoutNoKeyWords    = os.path.join(dataPath, titleKeywordsWithoutNoKeyWords)
abstractTitle                     = os.path.join(dataPath, abstractTitle)
abstractKeywordsWithNoKeyWords    = os.path.join(dataPath, abstractKeywordsWithNoKeyWords) 
abstractKeywordsWithoutNoKeyWords = os.path.join(dataPath, abstractKeywordsWithoutNoKeyWords)

print(titleKeywordsWithNoKeyWords)
print(titleKeywordsWithoutNoKeyWords)
print(abstractTitle)
print(abstractKeywordsWithNoKeyWords)
print(abstractKeywordsWithoutNoKeyWords)

#myfile = open('results.txt', 'w', encoding='utf-8')
fTitleKeywordsWithNoKeyWords       = open(titleKeywordsWithNoKeyWords,"w+", encoding='utf-8')
fTitleKeywordsWithoutNoKeyWords    = open(titleKeywordsWithoutNoKeyWords,"w+", encoding='utf-8')
fAbstractTitle                     = open(abstractTitle,"w+", encoding='utf-8')
fAbstractKeywordsWithNoKeyWords    = open(abstractKeywordsWithNoKeyWords,"w+", encoding='utf-8')
fAbstractKeywordsWithoutNoKeyWords = open(abstractKeywordsWithoutNoKeyWords,"w+", encoding='utf-8')
```

```python
#for each actual paper link, open it and get title, author, abstract and keywords.
for actualPaperLink in actualPapersLinks:
    flag = True
    page = requests.get(actualPaperLink)
    soup = BeautifulSoup(page.content, 'html.parser')
    
    title = soup.find_all(id='title')[0].get_text()
    print(title)
    print()

    author = soup.find_all(id='author')[0].get_text()
    print(author)
    print()

    abstract = soup.find_all(id='abstract')[0].get_text()
    abstract = abstract.strip()
    abstract = abstract[10:]
    print(abstract)
    print()

    keywords = soup.find_all(id='paperSubject')
    if not keywords:
        print('<No Keywords>')
        keywords = '<No Keywords>'
        flag = False
    else:
        keywords = soup.find_all(id='paperSubject')[0].get_text()
        keywords = keywords.strip() 
        keywords = keywords[10:]
        print(keywords)

    print('---------------------------------------------------------------')
    #f.write(foo.encode('utf8'))
    str1 = title    + '\t' + keywords + '\n'    
    str2 = abstract + '\t' + title    + '\n'
    str3 = abstract + '\t' + keywords + '\n'
    
    fTitleKeywordsWithNoKeyWords.write(str1)
    if(flag == True):
        fTitleKeywordsWithoutNoKeyWords.write(str1)
        fAbstractKeywordsWithoutNoKeyWords.write(str3)
        
    fAbstractTitle.write(str2)
    fAbstractKeywordsWithNoKeyWords.write(str3) 

fTitleKeywordsWithNoKeyWords.close()
fTitleKeywordsWithoutNoKeyWords.close()
fAbstractTitle.close()
fAbstractKeywordsWithNoKeyWords.close()
fAbstractKeywordsWithoutNoKeyWords.close()

print('Done')
```
